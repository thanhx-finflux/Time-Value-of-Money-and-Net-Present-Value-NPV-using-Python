Time Value of Money (TVM) is a financial concept that recognizes the value of money changes over time due to its potential earning capacity. This principle is based on the idea that a specific amount of money today is worth more than the same amount in the future because of its ability to earn interest or investment returns.

NPV (Net Present Value) is a method used to evaluate the profitability of an investment or project by calculating the present value of expected future cash flows, discounted back to their value today. NPV helps in determining whether an investment will yield a positive return after accounting for the time value of money.

### Key Concepts of TVM and NPV:
1. **Present Value (PV)**: The current value of a future sum of money or stream of cash flows given a specified rate of return. It answers the question, "How much is a future amount worth today?"
2. **Future Value (FV)**: The value of a current asset at a specified date in the future based on an assumed rate of growth. It answers the question, "How much will a current amount be worth in the future?"
3. **Discount Rate**: The interest rate used to discount future cash flows to their present value. It reflects the opportunity cost of capital, risk, and inflation.
4. **Cash Flows**: The amounts of money being transferred into and out of a business, project, or investment over time.
5. **NPV Calculation**: The formula for NPV is:

   $$
   NPV = \sum \left( \frac{C_t}{(1 + r)^t} \right) - C_0
   $$

   Where:

   - $C_t$ = Cash flow at time t
   - $r$ = Discount rate
   - $t$ = Time period
   - $C_0$ = Initial investment
   - $\sum$ = Summation of all cash flows over the investment period

### Importance of TVM and NPV:
- **Investment Decisions**: NPV helps investors and businesses decide whether to undertake a project or investment by comparing the present value of cash inflows to the initial investment.
- **Risk Assessment**: By incorporating the discount rate, NPV accounts for the risk associated with future cash flows.
- **Resource Allocation**: Helps in prioritizing projects based on their profitability and alignment with financial goals.
- **Financial Planning**: Assists in long-term financial planning by evaluating the impact of time on money.
Understanding TVM and NPV is crucial for making informed financial decisions, optimizing investments, and maximizing returns over time.

```python
# Time Value of Money (TVM) and Net Present Value (NPV) Calculator
# This script calculates the future value of an investment and the net present value of a series of cash flows.
import numpy_financial as npf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example usage:# Future Value Calculation
initial_investment = 1000  # Initial investment amount
annual_rate = 0.05         # Annual interest rate (5%)
years = 10                  # Number of years
future_value = initial_investment * (1 + annual_rate) ** years
print(f"Future Value after {years} years: ${future_value:.2f}")
```

```python
# After getting the future value, we can also calculate the Net Present Value (NPV) of a series of cash flows.
#%pip install numpy-financial

cash_flows = [-1000, 300, 400, 500, 600]  # Initial investment followed by cash inflows
discount_rate = 0.05  # Discount rate (5%)
npv = npf.npv(discount_rate, cash_flows)
print(f"Net Present Value (NPV): ${npv:.2f}")
# Plotting the cash flows
years = np.arange(len(cash_flows))
plt.bar(years, cash_flows, color='skyblue')
plt.xlabel('Years')
plt.ylabel('Cash Flow ($)')
plt.title('Cash Flows Over Time')
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
```

```python
# Scenario Analysis 1: Varying the discount rate to see its effect on NPV
discount_rates = np.linspace(0.01, 0.15, 100)
npvs = [npf.npv(rate, cash_flows) for rate in discount_rates]
plt.plot(discount_rates, npvs, color='orange')
plt.xlabel('Discount Rate')
plt.ylabel('Net Present Value (NPV) ($)')
plt.title('NPV vs Discount Rate')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid()
plt.show()
```

Discount rates can vary based on factors such as inflation, risk, and opportunity cost. A higher discount rate typically reduces the present value of future cash flows, making investments less attractive. Conversely, a lower discount rate increases the present value, making investments more appealing.

```python
# Scenario Analysis 2: Varying the annual interest rate to see its effect on future value
annual_rates = np.linspace(0.01, 0.15, 100)
future_values = [initial_investment * (1 + rate) ** years for rate in annual_rates]
plt.plot(annual_rates, future_values, color='green')
plt.xlabel('Annual Interest Rate')
plt.ylabel('Future Value ($)')
plt.title('Future Value vs Annual Interest Rate')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid()
plt.show()
```

```python
# Senario Analysis 3: Varying the number of years to see its effect on future value
years_range = np.arange(1, 31)
future_values_years = [initial_investment * (1 + annual_rate) ** year for year in years_range]
plt.plot(years_range, future_values_years, color='purple')
plt.xlabel('Number of Years')
plt.ylabel('Future Value ($)')
plt.title('Future Value vs Number of Years')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid()
plt.show()
```

```python
# Senario Analysis 4: Varying the initial investment to see its effect on future value
initial_investments = np.linspace(500, 5000, 100)
future_values_investment = [investment * (1 + annual_rate) ** years for investment in initial_investments]
plt.plot(initial_investments, future_values_investment, color='red')
plt.xlabel('Initial Investment ($)')
plt.ylabel('Future Value ($)')
plt.title('Future Value vs Initial Investment')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid()
plt.show()
```

```python
# Senario Analysis 5: Comparison of NPV for different cash flow scenarios
cash_flow_scenarios = {
    'Scenario 1': [-1000, 300, 400, 500, 600],
    'Scenario 2': [-1000, 200, 300, 400, 500],
    'Scenario 3': [-1000, 400, 500, 600, 700]
}
npvs_scenarios = {name: npf.npv(discount_rate, flows) for name, flows in cash_flow_scenarios.items()}
plt.bar(npvs_scenarios.keys(), npvs_scenarios.values(), color=['blue', 'orange', 'green'])
plt.xlabel('Cash Flow Scenarios')
plt.ylabel('Net Present Value (NPV) ($)')
plt.title('NPV Comparison Across Cash Flow Scenarios')
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
print(f'NPV Comparison: {npvs_scenarios}')
```

```python
# Senario Analysis 6: Comparison of Future Value for different initial investments
initial_investment_scenarios = [500, 1000, 2000, 3000]
for investment in initial_investment_scenarios:
    future_value = investment * (1 + annual_rate) ** years
    plt.plot(years_range, [investment * (1 + annual_rate) ** year for year in years_range], label=f'Initial: ${investment}')
plt.xlabel('Years')
plt.ylabel('Future Value ($)')
plt.title('Future Value Comparison Across Initial Investments')
plt.legend()
plt.show()
```

Future value (FV) comparison across different initial investments can be visualized using a line graph. Each line represents the growth of an initial investment over time at a constant interest rate. This allows for easy comparison of how different starting amounts grow over the same period.

```python
# Scenario Analysis 7: Assumption that initial investment requires $2000 with project will have an annual cash inflow of $1000 over 3 years.
# In order to compare the $2000 with $3000 in total, the analyst needs to consider the time value of money which is discounted at a rate of 5% per annum.
initial_investment_2000 = -2000
annual_cash_inflow = 1000
discount_rate = 0.05
cash_flows_2000 = [initial_investment_2000] + [annual_cash_inflow] * 3
npv_2000 = npf.npv(discount_rate, cash_flows_2000)
print(f"Net Present Value (NPV) for $2000 investment: ${npv_2000:.2f}")
initial_investment_3000 = -3000
cash_flows_3000 = [initial_investment_3000] + [annual_cash_inflow] * 3
npv_3000 = npf.npv(discount_rate, cash_flows_3000)
print(f"Net Present Value (NPV) for $3000 investment: ${npv_3000:.2f}")
# Plotting the cash flows for both scenarios
years = np.arange(len(cash_flows_2000))
plt.bar(years - 0.2, cash_flows_2000, width=0.4, label='$2000 Investment', color='skyblue')
plt.bar(years + 0.2, cash_flows_3000, width=0.4, label='$3000 Investment', color='orange')
plt.xlabel('Years')
plt.ylabel('Cash Flow ($)')
plt.title('Cash Flows Over Time for $2000 vs $3000 Investment')
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.show()
# Discount rates can vary based on factors such as inflation, risk, and opportunity cost. 
# A higher discount rate typically reduces the present value of future cash flows, 
# making investments less attractive. Conversely, a lower discount rate increases the present value, 
# making investments more appealing.
```

In this example, the NPV is positive, indicating that the investment is expected to generate more cash than the cost of the investment when considering the time value of money. This suggests that the $2000 investment is a worthwhile opportunity.

```python
# Scenario Analysis 8: Initial investment requires $1000 at time t and cash flows of of t+1 are $400, 450, 350, 475, 500, discounted at a rate of 10% per annum.
initial_investment_1000 = -1000
discount_rate_10 = 0.10
cash_flows_1000 = [initial_investment_1000, 400, 450, 350, 475, 500] # Time t=0 to t=5
npv_1000 = npf.npv(discount_rate_10, cash_flows_1000)
print(f"Net Present Value (NPV) for $1000 investment: ${npv_1000:.2f}")
# Plot the cash flows for this scenario
years_1000 = np.arange(len(cash_flows_1000))
plt.bar(years_1000, cash_flows_1000, color='lightgreen')
plt.xlabel('Years')
plt.ylabel('Cash Flow ($)')
plt.title('Cash Flows Over Time for $1000 Investment')
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
```
